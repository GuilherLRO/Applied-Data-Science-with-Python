Looking for an example? Here's what our course assistant put together for the **Ann Arbor, MI, USA** area using **sports and athletics** as the topic. [Example Solution File](./readonly/Assignment4_example.pdf)

```python
%matplotlib notebook

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker

plt.style.use('seaborn-notebook')

df_travel = pd.read_csv('travel_sevices.csv', skiprows=3);

#clean data 
df_travel.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis=1, inplace = True) 

years = np.arange(1960,1996,1);  
dropcols_travel = np.array([str(x) for x in years]); 

df_travel.drop(dropcols_travel, axis =1, inplace = True)

df_travel.set_index('Country Name', inplace = True); 

countries = ['Mexico', 'Peru', 'Chile', 'Argentina', 'Colombia']

#get data for the countries that we're interested in
df_travel = df_travel.loc[countries] 
df_travel = df_travel.reset_index()


#data.set_index('Country Name', inplace = True);
df_travel = df_travel.set_index(['Country Name'], drop=True)
data_travel = df_travel.T.set_index(np.arange(1996,2021,1))


#plot data
ax = data_travel.plot(figsize=(10,5), style=['-', '-', '-','-.']);


plt.xticks(np.arange(1996,2022,2))

plt.gca().spines['top'].set_visible(False) 
plt.gca().spines['right'].set_visible(False) 
ax.set_xticks(np.arange(1997,2022,2), minor=True)

#add a label to the x axis
plt.xlabel('Year')

#add a label to the y axis
plt.ylabel('% of travel service exports')

plt.axhline(y=0, color='#DCDCDC', linestyle=':') 
ax.set_title('Comparison of travel services % of Chile with Mexico, Peru, Colombia and Argentina over last 24 years (1996-2020)'); 
ttl = ax.title 
ttl.set_position([.5, 1.05])

#update Chile's line width 
lines = ax.get_lines() 
for l in lines: 
    if l.get_label() == 'Chile': 
        l.set_linewidth(3) 
    else : 
        l.set_linewidth(1) 
```
