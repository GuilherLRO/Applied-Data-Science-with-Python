# Week 4

```python
G = nx.read_edgelist('w4.txt',delimiter = '\t',create_using=nx.Graph) 
plt.figure(figsize=(5,4));
pos = nx.circular_layout(G);
plt.axis('off');

nx.draw_networkx(G,pos, alpha=0.7, with_labels=True, edge_color='.6')
```

```python
degrees = G.degree()
n = []
d = []
for i in degrees:
    n.append(i[0])
    d.append(i[1])

hist = [d.count(i) / float(len(d)) for i in set(d)]
```

```python
plt.bar(list(set(d)),hist)
plt.xlabel('Degree')
plt.ylabel('Fraction of nodes')
plt.show()
```

```python
G = nx.barabasi_albert_graph(100000,1)
degrees = G.degree()
n = []
d = []
for i in degrees:
    n.append(i[0])
    d.append(i[1])

hist = [d.count(i) / float(len(d)) for i in set(d)]
```

```python
plt.scatter(list(set(d)),hist)
plt.xlabel('Degree')
plt.ylabel('Fraction of nodes')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

```python
G1 = nx.watts_strogatz_graph(100000,6,0.04)
degrees = G1.degree()

n = []
d = []

for i in degrees:
    n.append(i[0])
    d.append(i[1])

hist = [d.count(i) / float(len(d)) for i in set(d)]

plt.bar(list(set(d)),hist)
plt.xlabel('Degree')
plt.ylabel('Fraction of nodes')
plt.show()
```
